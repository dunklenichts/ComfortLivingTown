```python
import pandas as pd
from googletrans import Translator, constants
from textblob import TextBlob
```

создание нового фрейма df_blob и перевод новостей на английский для использования модели определения тональностей библиотеки TextBlob

```python
df_blob = pd.DataFrame()

translator = Translator()

df_blob['news'] = all_news
df_blob['index'] = idxs
df_blob['en_news'] = df_blob['news'].apply(translator.translate, dest='en').apply(getattr, args=('text',))
en_list = df_blob['en_news'].tolist()
df_blob
```

изменение значений тональностей в диапазоне целых чисел от -1 до 1, а также избавление от столбца с англйиским переводом новостей

```python
sentiment_list = []

for new in en_list:
  analysis = TextBlob(new).sentiment
  if analysis.polarity < 0:
    sentiment_list.append(-1)
  elif (analysis.polarity <= 0.2):
    sentiment_list.append(0)
  elif (analysis.polarity > 0.2):
    sentiment_list.append(1)

df_blob = df_blob.drop('en_news', axis=1)
df_blob['sentiment'] = sentiment_list
df_blob
```

```python
from dostoevsky.tokenization import RegexTokenizer
from dostoevsky.models import FastTextSocialNetworkModel
import warnings
warnings.filterwarnings('ignore')
```

```python
tokenizer = RegexTokenizer()
# tokens = tokenizer.split('всё очень плохо')  # [('всё', None), ('очень', None), ('плохо', None)]

model = FastTextSocialNetworkModel(tokenizer=tokenizer)

results = model.predict(all_news, k=2)
sentiment_list = []
results = model.predict(all_news, k=2)
for sentiment in results:
    sentiment_list.append(sentiment)

df_dostoevsky = pd.DataFrame()
df_dostoevsky['news'] = all_news

neutral_list = []
negative_list = []
positive_list = []
speech_list = []
skip_list = []
for sentiment in sentiment_list:
    neutral = sentiment.get('neutral')
    negative = sentiment.get('negative')
    positive = sentiment.get('positive')
    if (neutral is None):
        neutral_list.append(0)
    else:
        neutral_list.append(sentiment.get('neutral'))
    if (negative is None):
        negative_list.append(0)
    else:
        negative_list.append(sentiment.get('negative'))
    if (positive is None):
        positive_list.append(0)
    else:
        positive_list.append(sentiment.get('positive'))

df_dostoevsky['index'] = idxs
df_dostoevsky['neutral'] = neutral_list
df_dostoevsky['negative'] = negative_list
df_dostoevsky['positive'] = positive_list

df_dostoevsky.loc[df_dostoevsky['positive'] > 0, 'transcription'] = 1
df_dostoevsky.loc[df_dostoevsky['negative'] > 0, 'transcription'] = -1
df_dostoevsky.loc[(df_dostoevsky['positive'] == 0) & (df_dostoevsky['negative'] == 0), 'transcription'] = 0

df_dostoevsky['transcription'] = df_dostoevsky['transcription'].astype(int)

df_dostoevsky
```

```python
ecology_neg = len(df_dostoevsky[(df_dostoevsky['transcription']==-1) & (df_dostoevsky['index'] == 'ecology')])
ecology_pos = len(df_dostoevsky[(df_dostoevsky['transcription']==1) & (df_dostoevsky['index'] == 'ecology')])
ecology_neu = len(df_dostoevsky[(df_dostoevsky['transcription']==0) & (df_dostoevsky['index'] == 'ecology')])

infrasturture_neg = len(df_dostoevsky[(df_dostoevsky['transcription']==-1) & (df_dostoevsky['index'] == 'infrastructure')])
infrasturture_pos = len(df_dostoevsky[(df_dostoevsky['transcription']==1) & (df_dostoevsky['index'] == 'infrastructure')])
infrasturture_neu = len(df_dostoevsky[(df_dostoevsky['transcription']==0) & (df_dostoevsky['index'] == 'infrastructure')])

house_neg = len(df_dostoevsky[(df_dostoevsky['transcription']==-1) & (df_dostoevsky['index'] == 'house')])
house_pos = len(df_dostoevsky[(df_dostoevsky['transcription']==1) & (df_dostoevsky['index'] == 'house')])
house_neu = len(df_dostoevsky[(df_dostoevsky['transcription']==0) & (df_dostoevsky['index'] == 'house')])

criminal_neg = len(df_dostoevsky[(df_dostoevsky['transcription']==-1) & (df_dostoevsky['index'] == 'criminal')])
criminal_pos = len(df_dostoevsky[(df_dostoevsky['transcription']==1) & (df_dostoevsky['index'] == 'criminal')])
criminal_neu = len(df_dostoevsky[(df_dostoevsky['transcription']==0) & (df_dostoevsky['index'] == 'criminal')])

town_neg = len(df_dostoevsky[(df_dostoevsky['transcription']==-1) & (df_dostoevsky['index'] == 'town')])
town_pos = len(df_dostoevsky[(df_dostoevsky['transcription']==1) & (df_dostoevsky['index'] == 'town')])
town_neu = len(df_dostoevsky[(df_dostoevsky['transcription']==0) & (df_dostoevsky['index'] == 'town')])

dff = {'positive' : pd.Series([ecology_pos, infrasturture_pos, house_pos, criminal_pos,
                              town_pos], index = ['экология', 'инфраструктура', 'ЖКХ',
                                                   'закон и порядок', 'городская активность']),
       'negative' : pd.Series([ecology_neg, infrasturture_neg, house_neg, criminal_neg,
                              town_neg], index = ['экология', 'инфраструктура', 'ЖКХ',
                                                   'закон и порядок', 'городская активность']),
       'neutral' : pd.Series([ecology_neu, infrasturture_neu, house_neu, criminal_neu,
                              town_neu], index = ['экология', 'инфраструктура', 'ЖКХ',
                                                   'закон и порядок', 'городская активность']),
       'total' : pd.Series([(ecology_neu + ecology_neg + ecology_pos),
                            (infrasturture_neu + infrasturture_neg + infrasturture_pos),
                            (house_neu + house_neg + house_pos),
                            (criminal_neu + criminal_neg + criminal_pos),
                            (town_neu + town_neg + town_pos)],
                           index = ['экология', 'инфраструктура', 'ЖКХ',
                                    'закон и порядок', 'городская активность'])

       }
df_temp = pd.DataFrame(dff)
df_temp
```

```python
df_precent = pd.DataFrame(dff)

df_precent['precent_pos'] = df_precent['positive'] / df_precent['total'] * 100
df_precent['precent_neg'] = df_precent['negative'] / df_precent['total'] * 100
df_precent['precent_neu'] = df_precent['neutral'] / df_precent['total'] * 100

# df_precent.dtypes
df_precent['precent_neu'] = df_precent['precent_neu'].fillna (0)
df_precent['precent_neg'] = df_precent['precent_neg'].fillna (0)
df_precent['precent_pos'] = df_precent['precent_pos'].fillna (0)

df_precent['precent_neu'] = df_precent['precent_neu'].astype(int)
df_precent['precent_neg'] = df_precent['precent_neg'].astype(int)
df_precent['precent_pos'] = df_precent['precent_pos'].astype(int)

df_precent = df_precent.drop('positive', axis=1)
df_precent = df_precent.drop('negative', axis=1)
df_precent = df_precent.drop('total', axis=1)
df_precent = df_precent.drop('neutral', axis=1)

df_precent
```
